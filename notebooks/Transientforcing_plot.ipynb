```python
directory = '../Estimates/Transient_forcing_estimates/'
models = [ f.name for f in os.scandir(directory) if f.is_dir() and f.name !='.ipynb_checkpoints']
models.sort()
models;
```

```python
importlib.reload(f)
experiments = ['1pctCO2', 'historical', 'ssp126', 'ssp245', 'ssp370', 'ssp585']

explengths = {'1pctCO2':150, 'historical':165, 'ssp126':86, 'ssp245':86, 'ssp370':86, 'ssp585':86}
startyears = {'1pctCO2':1850, 'historical':1850, 'ssp126':2015, 'ssp245':2015, 'ssp370':2015, 'ssp585':2015}
nrows = len(experiments)

fig = plt.figure(figsize = [18,20]);
plt.subplots_adjust(hspace=0.2, wspace = 0.1)
outer = gridspec.GridSpec(nrows, 1)

# assign one color to each model:
palette = sns.color_palette(None, len(models))

for (k,exp) in enumerate(experiments):
    print(exp)
    gs_k = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec = outer[k], hspace = 0.1)
    f_plot = fig.add_subplot(gs_k[0])
    t_plot = fig.add_subplot(gs_k[1])

    f_plot.tick_params(axis='both',labelsize=14); t_plot.tick_params(axis='both',labelsize=14)  
    f_plot.set_ylabel(exp, fontsize=14, fontweight = 'bold')
    #t_plot.set_ylabel('T(t) [°C]', fontsize=14)
    t_plot.axhline(0, color='k', linewidth = 0.5) # horizontal lines
    if exp == experiments[0]: # top row
        f_plot.set_title('Effective radiative forcing [$W/m^2$]',fontsize = 18)
        t_plot.set_title('Temperature difference [°C]',fontsize = 18)
        
        # include also etminan co2 forcing for 1pctCO2
        concfile = pd.read_excel("../Estimates/meinshausen_1pctCO2concentrations.xlsx", header=2)
        f_plot.plot(concfile['v YEARS/GAS >'].values, f.etminan_co2forcing(concfile['CO2']).values, color = 'black', zorder = 99, linewidth = 2)
    
    if exp[:3] == 'ssp':
        startind = 165
        expstr = 'historicaland' + exp
        f_plot.set_ylim(0,12)
    else:
        startind = 0
        expstr = exp
    years = np.arange(startyears[exp], startyears[exp]+explengths[exp])
    f_plot.set_xlim(years[0],years[-1]);# f_plot.set_ylim(-2.5,6.5)
    t_plot.set_xlim(years[0],years[-1]);# t_plot.set_ylim(-0.5,0.8)
    
    for (i,model) in enumerate(models):
    #for (i,model) in enumerate(['CNRM-CM6-1']):
        #print(model)
        modeldirectory = os.path.join(directory, model)
        modelexp_names = [ f.name for f in os.scandir(modeldirectory) if f.is_dir() and f.name !='.ipynb_checkpoints']
        if exp in modelexp_names:
            members = f.find_members(model, exp, datatype = 'forcing')
            members = list(dict.fromkeys(members)) # drop duplicates
        else:
            continue

        model_ERFs = np.full((explengths[exp], len(members)), np.nan)
        model_tas_diffs = np.full((explengths[exp], len(members)), np.nan)
        for (m, member) in enumerate(members): # plot very thin lines for individual members
            if model == 'KACE-1-0-G' and exp == 'historical' and member[:2] == 'r3':
                # something is wrong about these numbers, need to look into it
                continue
            modelfilename = model + '_' + expstr + '_' + member + '_ERF.txt'
            tas_filename = model + '_' + expstr + '_' + member + '_tas_diff.txt'
            modelexpdir = os.path.join(directory, model, exp)
            ERF = pd.read_table(modelexpdir + '/' + modelfilename,index_col=0).values[startind:(startind+explengths[exp])]
            if len(ERF) != explengths[exp]:
                print(model, exp, member, len(ERF))
            tas_diff = pd.read_table(modelexpdir + '/' + tas_filename,index_col=0).values[startind:(startind+explengths[exp])]
            
            f_plot.plot(years[:len(ERF)], ERF, linewidth = 0.1, color = palette[i])
            t_plot.plot(years[:len(tas_diff)], tas_diff, linewidth = 0.1, color = palette[i])
            
            model_ERFs[:len(ERF), m] = np.squeeze(ERF)
            model_tas_diffs[:len(tas_diff), m] = np.squeeze(tas_diff)
        
        # plot thicker lines for mean of members (also models with only one member)
        member_mean_ERF = np.nanmean(model_ERFs, axis=1)
        member_mean_tas_diff = np.nanmean(model_tas_diffs, axis=1)
        
        f_plot.plot(years[:len(member_mean_ERF)], member_mean_ERF, linewidth = 0.7, color = palette[i], label = model)
        t_plot.plot(years[:len(member_mean_tas_diff)], member_mean_tas_diff, linewidth = 0.7, color = palette[i])
    
    if exp[:3] == 'ssp':
        forcinglevel = float(exp[4:])/10
        f_plot.arrow(2103, forcinglevel, -2, 0, color = "black",\
                            clip_on = False, head_width = 0.8)

    #plt.savefig('../Figures/transient_forcings_plot.pdf', format='pdf', dpi=600, bbox_inches="tight")
```
