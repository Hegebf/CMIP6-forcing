# Plot transient forcing estimates

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import importlib
import my_functions as f
importlib.reload(f); # needed to load new changes in my_functions
import matplotlib.gridspec as gridspec
```

```python
directory = '../Estimates/Transient_forcing_estimates/'
models = [ f.name for f in os.scandir(directory) if f.is_dir() and f.name !='.ipynb_checkpoints']
models.sort()
models.remove('CNRM-ESM2-1') # only if plotting the ensemble, because these values are known outliers
models.remove('CIESM')
models.remove('CMCC-CM2-SR5')
models
```

```python
importlib.reload(f)
experiments = ['1pctCO2', 'historical', 'ssp126', 'ssp245', 'ssp370', 'ssp585']

explengths = {'1pctCO2':150, 'historical':165, 'ssp126':86, 'ssp245':86, 'ssp370':86, 'ssp585':86}
startyears = {'1pctCO2':1850, 'historical':1850, 'ssp126':2015, 'ssp245':2015, 'ssp370':2015, 'ssp585':2015}
nrows = len(experiments)

fig = plt.figure(figsize = [18,20]);
plt.subplots_adjust(hspace=0.2, wspace = 0.1)
outer = gridspec.GridSpec(nrows, 1)

# assign one color to each model:
palette = sns.color_palette(None, len(models))

for (k,exp) in enumerate(experiments):
    print(exp)
    gs_k = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec = outer[k], hspace = 0.1)
    f_plot = fig.add_subplot(gs_k[0])
    t_plot = fig.add_subplot(gs_k[1])

    f_plot.tick_params(axis='both',labelsize=14); t_plot.tick_params(axis='both',labelsize=14)  
    f_plot.set_ylabel(exp, fontsize=14, fontweight = 'bold')
    #t_plot.set_ylabel('T(t) [°C]', fontsize=14)
    t_plot.axhline(0, color='k', linewidth = 0.5) # horizontal lines
    if exp == experiments[0]: # top row
        f_plot.set_title('Effective radiative forcing [$W/m^2$]',fontsize = 18)
        t_plot.set_title('Temperature difference [°C]',fontsize = 18)
        
        # include also etminan co2 forcing for 1pctCO2
        concfile = pd.read_excel("../Estimates/meinshausen_1pctCO2concentrations.xlsx", header=2)
        f_plot.plot(concfile['v YEARS/GAS >'].values, f.etminan_co2forcing(concfile['CO2']).values, color = 'black', zorder = 99, linewidth = 2)
    
    if exp[:3] == 'ssp':
        startind = 165
        expstr = 'historicaland' + exp
        f_plot.set_ylim(0,12)
    else:
        startind = 0
        expstr = exp
    years = np.arange(startyears[exp], startyears[exp]+explengths[exp])
    f_plot.set_xlim(years[0],years[-1]);# f_plot.set_ylim(-2.5,6.5)
    t_plot.set_xlim(years[0],years[-1]);# t_plot.set_ylim(-0.5,0.8)
    
    for (i,model) in enumerate(models):
    #for (i,model) in enumerate(['CNRM-CM6-1']):
        #print(model)
        modeldirectory = os.path.join(directory, model)
        modelexp_names = [ f.name for f in os.scandir(modeldirectory) if f.is_dir() and f.name !='.ipynb_checkpoints']
        if exp in modelexp_names:
            members = f.find_members(model, exp, datatype = 'forcing')
            members = list(dict.fromkeys(members)) # drop duplicates
        else:
            continue

        model_ERFs = np.full((explengths[exp], len(members)), np.nan)
        model_tas_diffs = np.full((explengths[exp], len(members)), np.nan)
        for (m, member) in enumerate(members): # plot very thin lines for individual members
            modelfilename = model + '_' + expstr + '_' + member + '_ERF.txt'
            tas_filename = model + '_' + expstr + '_' + member + '_tas_diff.txt'
            modelexpdir = os.path.join(directory, model, exp)
            ERF = pd.read_table(modelexpdir + '/' + modelfilename,index_col=0).values[startind:(startind+explengths[exp])]
            if len(ERF) != explengths[exp]:
                print(model, exp, member, len(ERF))
            tas_diff = pd.read_table(modelexpdir + '/' + tas_filename,index_col=0).values[startind:(startind+explengths[exp])]
            
            f_plot.plot(years[:len(ERF)], ERF, linewidth = 0.1, color = palette[i])
            t_plot.plot(years[:len(tas_diff)], tas_diff, linewidth = 0.1, color = palette[i])
            
            model_ERFs[:len(ERF), m] = np.squeeze(ERF)
            model_tas_diffs[:len(tas_diff), m] = np.squeeze(tas_diff)
            
            # print out info about data that potentially contain errors:
            #if exp == 'ssp126' and np.abs(tas_diff[0])>0.5:
            #if exp == 'ssp585' and np.abs(tas_diff[-1])>0.5:
            #    print(model, exp, member, 'check branch info')
                
        # plot thicker lines for mean of members (also models with only one member)
        member_mean_ERF = np.nanmean(model_ERFs, axis=1)
        member_mean_tas_diff = np.nanmean(model_tas_diffs, axis=1)
        
        f_plot.plot(years[:len(member_mean_ERF)], member_mean_ERF, linewidth = 0.7, color = palette[i], label = model)
        t_plot.plot(years[:len(member_mean_tas_diff)], member_mean_tas_diff, linewidth = 0.7, color = palette[i])
        
        
        
    if exp[:3] == 'ssp':
        forcinglevel = float(exp[4:])/10
        f_plot.arrow(2103, forcinglevel, -2, 0, color = "black",\
                            clip_on = False, head_width = 0.8)

    #plt.savefig('../Figures/transient_forcings_plot.pdf', format='pdf', dpi=600, bbox_inches="tight")
```

## Plot shading for both CMIP6 and CMIP5 models:

```python
rmse_limit = np.inf #0.15

importlib.reload(f)
experiments = ['1pctCO2', 'historical', 'ssp126', 'ssp245', 'ssp370', 'ssp585']
explengths = {'1pctCO2':150, 'historical':165, 'ssp126':86, 'ssp245':86, 'ssp370':86, 'ssp585':86}
startyears = {'1pctCO2':1850, 'historical':1850, 'ssp126':2015, 'ssp245':2015, 'ssp370':2015, 'ssp585':2015}
nrows = len(experiments)

fig = plt.figure(figsize = [18,20]);
plt.subplots_adjust(hspace=0.2, wspace = 0.1)
outer = gridspec.GridSpec(nrows, 1)

cmip5directory = '../Estimates/CMIP5_forcing_estimates/'
tas_diff_rmse = {}
cmip5_1pctCO2_rmse = {}
cmip6models_included = []
cmip5models_included = []

for (k,exp) in enumerate(experiments):
    print(exp)
    gs_k = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec = outer[k], hspace = 0.1)
    f_plot = fig.add_subplot(gs_k[0])
    t_plot = fig.add_subplot(gs_k[1])
    f_plot.tick_params(axis='both',labelsize=14); t_plot.tick_params(axis='both',labelsize=14)  
    f_plot.set_ylabel(exp, fontsize=14, fontweight = 'bold')
    #t_plot.set_ylabel('T(t) [°C]', fontsize=14)
    t_plot.axhline(0, color='k', linewidth = 0.5) # horizontal lines
    if exp == experiments[0]: # top row
        f_plot.set_title('Effective radiative forcing [$W/m^2$]',fontsize = 18)
        t_plot.set_title('Temperature difference [°C]',fontsize = 18)
    if exp[:3] == 'ssp':
        startind = 165
        expstr = 'historicaland' + exp
        f_plot.set_ylim(0,12)
    else:
        startind = 0
        expstr = exp
    years = np.arange(startyears[exp], startyears[exp]+explengths[exp])
    f_plot.set_xlim(np.min([years[0], 2006]),years[-1]);# f_plot.set_ylim(-2.5,6.5)
    t_plot.set_xlim(np.min([years[0], 2006]),years[-1]);# t_plot.set_ylim(-0.5,0.8)
    
    model_ERF = np.full((explengths[exp], len(models)), np.nan);
    model_tas_diff = np.full((explengths[exp], len(models)), np.nan);
    tas_diff_rmse[exp] = {}
    for (i,model) in enumerate(models):
        modeldirectory = os.path.join(directory, model)
        modelexp_names = [ f.name for f in os.scandir(modeldirectory) if f.is_dir() and f.name !='.ipynb_checkpoints']
        if exp in modelexp_names:
            members = f.find_members(model, exp, datatype = 'forcing')
            members = list(dict.fromkeys(members)) # drop duplicates
        else:
            continue

        member_ERFs = np.full((explengths[exp], len(members)), np.nan)
        member_tas_diffs = np.full((explengths[exp], len(members)), np.nan)
        for (m, member) in enumerate(members): # plot very thin lines for individual members
            modelfilename = model + '_' + expstr + '_' + member + '_ERF.txt'
            tas_filename = model + '_' + expstr + '_' + member + '_tas_diff.txt'
            modelexpdir = os.path.join(directory, model, exp)
            ERF = pd.read_table(modelexpdir + '/' + modelfilename,index_col=0).values[startind:(startind+explengths[exp])]
            tas_diff = pd.read_table(modelexpdir + '/' + tas_filename,index_col=0).values[startind:(startind+explengths[exp])]
            member_ERFs[:len(ERF), m] = np.squeeze(ERF)
            member_tas_diffs[:len(tas_diff), m] = np.squeeze(tas_diff)
                
        # plot thicker lines for mean of members (also models with only one member)
        tas_diff_mean = np.nanmean(member_tas_diffs, axis=1)
        tas_diff_rmse[exp][model] = np.sqrt(np.mean(tas_diff_mean**2))
        
        if tas_diff_rmse['1pctCO2'][model]<rmse_limit:
            model_ERF[:len(member_ERFs), i] = np.nanmean(member_ERFs, axis=1)
            model_tas_diff[:len(member_tas_diffs), i] = tas_diff_mean
            # otherwise the values remain nan
            if exp == '1pctCO2':
                cmip6models_included.append(model)
            

    model_ERF_max = np.nanmax(model_ERF, axis=1)
    model_ERF_min = np.nanmin(model_ERF, axis=1)
    model_ERF_mean = np.nanmean(model_ERF, axis=1)
    
    f_plot.fill_between(years, model_ERF_min, model_ERF_max, facecolor="gray",alpha=0.2)
    f_plot.plot(years, model_ERF_mean, color = 'black')
    
    model_tas_diff_max = np.nanmax(model_tas_diff, axis=1)
    model_tas_diff_min = np.nanmin(model_tas_diff, axis=1)
    model_tas_diff_mean = np.nanmean(model_tas_diff, axis=1)
    
    t_plot.fill_between(years, model_tas_diff_min, model_tas_diff_max, facecolor="gray",alpha=0.2)
    t_plot.plot(years, model_tas_diff_mean, color = 'black')
    
    if exp[:3] == 'ssp':
        forcinglevel = float(exp[4:])/10
        f_plot.arrow(2103, forcinglevel, -2, 0, color = "black",\
                            clip_on = False, head_width = 0.8)
    
    # corresponding CMIP5 experiments:
    if exp == '1pctCO2':
        cmip5filename = 'forcing_estimates_1pctCO2.txt'
        cmip5tasdiff_files = 'onepctCO2_tas_diffs.txt'
    elif exp == 'historical' or exp == 'ssp585': # the rcp files contain the same historical forcing
        cmip5filename = 'historical_and_rcp85_forcing_estimates.txt'
        cmip5tasdiff_files = 'historical_and_rcp85_tas_diffs.txt'
    elif exp == 'ssp126':
        cmip5filename = 'historical_and_rcp26_forcing_estimates.txt'  
        cmip5tasdiff_files = 'historical_and_rcp26_tas_diffs.txt'
    elif exp == 'ssp245':
        cmip5filename = 'historical_and_rcp45_forcing_estimates.txt'
        cmip5tasdiff_files = 'historical_and_rcp45_tas_diffs.txt'
    elif exp == 'ssp370':
        continue # no corresponding rcp experiment

    cmip5forcing = pd.read_table(cmip5directory + cmip5filename, index_col=0, sep = '\t', header = 0, na_values = '-')
    cmip5models = list(cmip5forcing)[1:]
    years = cmip5forcing['year'].values
    cmip5tas_diffs = pd.read_table(cmip5directory + cmip5tasdiff_files, index_col=0, sep = ',', header = 0, na_values = '')
    
    if exp == 'historical': # cmip5 historical period ends in 2005
        years = years[:156]
        cmip5forcing = cmip5forcing.iloc[:156,:]
        cmip5tas_diffs = cmip5tas_diffs.iloc[:156,:]
    elif exp[:3] == 'ssp':
        years = years[-95:]
        cmip5forcing = cmip5forcing.iloc[-95:,:].reset_index()
        cmip5tas_diffs = cmip5tas_diffs.iloc[-95:,:].reset_index()
    elif exp == '1pctCO2':
        years = years + startyears['1pctCO2']
        for model in cmip5models: 
            model_tas_diff = cmip5tas_diffs[model].values
            cmip5_1pctCO2_rmse[model] = np.sqrt(np.mean(model_tas_diff**2))
            if cmip5_1pctCO2_rmse[model] < rmse_limit:
                cmip5models_included.append(model)

    ERFtable = pd.DataFrame(np.full(np.shape(cmip5forcing[cmip5models]), np.nan), columns = cmip5models)
    tasdifftable = ERFtable.copy(deep=True)
    for model in cmip5models:
        if cmip5_1pctCO2_rmse[model]<rmse_limit:
            if exp == '1pctCO2' and model[:8] == 'GFDL-ESM':
                # CO2 increase stops after doubling in these two models
                continue
            ERFtable[model] = cmip5forcing[model]
            tasdifftable[model] = cmip5tas_diffs[model]
            
    
    model_ERF_max = np.nanmax(ERFtable.values, axis=1); 
    model_ERF_min = np.nanmin(ERFtable.values, axis=1)
    model_ERF_mean = np.nanmean(ERFtable.values, axis=1)
    
    f_plot.fill_between(years, model_ERF_min, model_ERF_max, facecolor="red",alpha=0.2)
    f_plot.plot(years, model_ERF_mean, color = 'red')
    
    model_tasdiff_max = np.nanmax(tasdifftable.values, axis=1);
    model_tasdiff_min = np.nanmin(tasdifftable.values, axis=1);
    model_tasdiff_mean = np.nanmean(tasdifftable.values, axis=1);
        
    t_plot.fill_between(years, model_tasdiff_min, model_tasdiff_max, facecolor="red",alpha=0.2)
    t_plot.plot(years, model_tasdiff_mean, color = 'red')
    
    
    #plt.savefig('../Figures/transient_forcings_shading_cmip5vs6_rmselim0p15.pdf', format='pdf', dpi=600, bbox_inches="tight")
    #plt.savefig('../Figures/transient_forcings_shading_cmip5vs6.pdf', format='pdf', dpi=600, bbox_inches="tight")

```
