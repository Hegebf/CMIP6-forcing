# Compare CMIP5 and CMIP6 abrupt-4xCO2 forcing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 # load parameters:

####### CMIP5 models #######:
filename = 'best_estimated_parameters_cmip5.txt'
parameter_table = pd.read_csv('../Estimates/CMIP5_forcing_estimates/' + filename, index_col=0, sep = '\t')
cmip5_models = parameter_table.index
F4x_3lambda_5 = 2*parameter_table['F2x']    
F4x_1lambda_5 = 2*parameter_table['GregoryF2x']
                              
####### CMIP6 models #######:
filename = 'best_estimated_parameters_allmembers4xCO2.csv'
parameter_table = pd.read_csv('../Estimates/' + filename, index_col=0)
cmip6_models = list(parameter_table.index)
F4x_3lambda_6 = 2*parameter_table['F2x']        
F4x_1lambda_6 = 2*parameter_table['GregoryF2x']
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (16,6))

for axis in ax:
    axis.tick_params(axis='both',labelsize=16)
    axis.set_xlabel('ERF [W/m$^2$]', fontsize = 16)
    axis.set_xlim(5,11)
    axis.set_ylabel('Number of models', fontsize = 16)
ax[0].hist(F4x_1lambda_6, alpha = 0.8, label = 'CMIP6'); ax[0].axvline(F4x_1lambda_6.mean(), color = 'C0', linewidth = 3)
ax[0].hist(F4x_1lambda_5, alpha = 0.8, label = 'CMIP5'); ax[0].axvline(F4x_1lambda_5.mean(), color = 'C1', linewidth = 3)
ax[0].legend(fontsize = 16, loc = 'upper right')
ax[0].set_title('1-$\lambda$ abrupt-4xCO2 ERF', fontsize = 16, fontweight = 'bold')

ax[1].hist(F4x_3lambda_6, alpha = 0.8); ax[1].axvline(F4x_3lambda_6.mean(), color = 'C0', linewidth = 3)
ax[1].hist(F4x_3lambda_5, alpha = 0.8); ax[1].axvline(F4x_3lambda_5.mean(), color = 'C1', linewidth = 3)
ax[1].set_title('3-$\lambda$ abrupt-4xCO2 ERF', fontsize = 16, fontweight = 'bold');

#plt.savefig('../Figures/comparing_4xCO2_forcing.pdf', format='pdf', dpi=600, bbox_inches="tight")
```
